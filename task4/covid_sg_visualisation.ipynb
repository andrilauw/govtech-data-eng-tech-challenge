```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
import requests
```

```python
country = "singapore"
start_time = "2020-03-01T00:00:00Z"
end_time = "2022-08-01T00:00:00Z"

URL = f"https://api.covid19api.com/country/{country}/status/confirmed?from={start_time}&to={end_time}"
```

```python
response = requests.get(URL)
```

```python
result = response.json()
```

```python
dates_list = []
cases_list = []
```

```python
for item in result:
    dates_list.append(item["Date"])
    cases_list.append(item["Cases"])
```

```python
xpoints = np.array(dates_list)
ypoints = np.array(cases_list)
```

```python
fig, ax = plt.subplots()
ax.plot(xpoints, ypoints)
plt.title('Total Covid-19 Cases in Singapore', fontsize=20)
plt.xlabel('Confirmed Date', fontsize=16)
plt.ylabel('Total Cases', fontsize=16)
start_y, end_y = ax.get_ylim()
ax.yaxis.set_ticks(np.arange(0, end_y, 100000))
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.0f'))
start_x, end_x = ax.get_xlim()
ax.xaxis.set_ticks(np.arange(start_x, end_x, 50))
plt.xticks(rotation=90)
plt.grid()
plt.show()
```
